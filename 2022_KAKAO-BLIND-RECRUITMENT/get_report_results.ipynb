## **Solution.py**

```python
from pprint import pprint
```

```python
from collections import defaultdict, Counter

def solution(id_list, report, k):
    rpt_dict = defaultdict(set)
    rpt_list = list()
    
    for rpt in set(report):
        id_, reported = rpt.split()
        rpt_list.append(reported)
        rpt_dict[id_].add(reported)
    res = set([rpt_id for rpt_id, rpt_num in Counter(rpt_list).items() if rpt_num>=k])
    return [len(rpt_dict[id_]&res) for id_ in id_list]
```

```python
inputs = [
    (["muzi", "frodo", "apeach", "neo"],["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"],2),
    (["con", "ryan"],["ryan con", "ryan con", "ryan con", "ryan con"],3)
]
outputs = [
    [2,1,1,0],
    [0,0]
]
```

```python
for idx, (i, o) in enumerate(zip(inputs, outputs)):
    sol_ = solution(*i)
    print(sol_==o, sol_, o)
```
